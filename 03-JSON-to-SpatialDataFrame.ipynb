## JSON to Spatial Dataframes
While CSV files are useful formats for storing data in a generic text-based formats, [JSON](https://developers.squarespace.com/what-is-json) formats are more flexible and hence are gaining traction in the client-server exchange of data (e.g. via REST requests). And because JSON - and GeoJSON, its geospatial derivative - are so pervasive in these cloud GIS computing, it's good to view how to convert these formats into spatial dataframes. 

Again, we'll examine this from both a Geopandas (via Fiona) and ArcGIS Python API perspective.
