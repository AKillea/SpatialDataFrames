## Shapefile to spatial dataframe - Geopandas
* http://geopandas.org/io.html

```python
import geopandas as gp
%matplotlib inline
```

```python
theShapefile = '../../data/HUC12.shp'
sdfHUC12 = gp.read_file(theShapefile)
type(sdfHUC12)
```

```python
sdfHUC12.describe()
```

```python
#Display as simple polygon FC
sdfHUC12.plot();
```

```python
#Display area as continuous values
sdfHUC12.plot('ACRES',cmap='GnBu');
```

```python
#Display as chorplet map using HUC8 attribute
sdfHUC12.plot('HUC_8',cmap='Set2');
```
